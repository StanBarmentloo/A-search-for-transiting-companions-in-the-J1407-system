```python
#imports
from astropy.io import ascii
import matplotlib.pyplot as plt
import matplotlib
```

```python
plt.style.use('seaborn-dark-palette')
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 20}
matplotlib.rc('font', **font)
```

```python
#reading in the data
data = ascii.read('Final_KARPA_Data_BRO2.csv')
time = data['col1']
flux = data['col2']
eflux = data['col3']

error_mask = eflux < 0.05 #remove values with errors larger than 5%

time = time[error_mask]
flux = flux[error_mask]
eflux = eflux[error_mask]
print(len(time))
```

```python
#creating the plot
plt.errorbar(time, flux, eflux, fmt='.', ms=3, elinewidth = 0.3)
plt.xlabel('MJD (Days)')
plt.ylabel('Normalised Flux')
plt.axhline(y = 1.0, linestyle = '--', c='black')
figure = plt.gcf()
figure.set_size_inches(18, 10)
#plt.savefig('KARPA_lightcurve.pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astropy.io.votable import parse
from astropy.io import ascii
```

```python
wasp = parse("WASP.SuperWASP.SW.xml").get_first_table()
tess = parse("TESS.TESS.Red.xml").get_first_table()
asas_g = parse("SLOAN.SDSS.g.xml").get_first_table()
kelt = ascii.read('kelt_manual.dat')
prompt_v = ascii.read('Bessel_V-1.txt') 
```

```python
font = {'family':'normal', 'weight':'normal', 'size':28}
plt.rc('font', **font)
plt.style.use('seaborn-dark-palette')

fig, ax = plt.subplots(1,1, figsize=(10,4))
#plt.plot(wasp.array['Wavelength'],wasp.array['Transmission']/np.max(wasp.array['Transmission']),label='WASP')
plt.plot(tess.array['Wavelength'],tess.array['Transmission']/np.max(tess.array['Transmission']),label='TESS', c= 'r')
plt.plot(asas_g.array['Wavelength'],asas_g.array['Transmission']/np.max(asas_g.array['Transmission']),label='ASAS & ASASSN',c= 'b')
plt.plot(kelt['wlen']*10., kelt['tx']/np.max(kelt['tx']),label='KELT',c= 'orange')
plt.plot(prompt_v['wlen']*10., prompt_v['tx']/np.max(prompt_v['tx']),label='PROMPT & ROAD',c= 'green')

plt.legend(loc = 'upper right', fontsize = 16)
plt.xlabel("Wavelength (Ångströms)")
plt.ylabel('Transmission')
plt.xlim(3500,11000)
fig = plt.gcf()
fig.set_size_inches(15, 7.5)
plt.savefig('Figure1.pdf')
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,4))
plt.plot(wasp.array['Wavelength'],wasp.array['Transmission'],label='WASP')
plt.plot(tess.array['Wavelength'],tess.array['Transmission'],label='TESS')
plt.plot(asas_g.array['Wavelength'],asas_g.array['Transmission'],label='ASAS g')
plt.plot(kelt['wlen']*10., kelt['tx'],label='KELT')
plt.plot(prompt_v['wlen']*10., prompt_v['tx']/100,label='PROMPT V')

plt.legend()
plt.xlabel('Wavelength [Angstroms]')
plt.ylabel('transmission')
plt.title('Filters observing J1407')
plt.xlim(3500,11000)
```
